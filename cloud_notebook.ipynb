Data Description and Visualization for a Airline Dataset

```python
import matplotlib.pyplot as plt
import pandas as pd
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv("airline.csv")
df.describe()
```

```python
# 1. Histogram of Available Seats per Week
plt.figure(figsize=(10, 6))
df['avail_seat_km_per_week'].hist(bins=20, edgecolor='black')
plt.title('Distribution of Available Seats per Week')
plt.xlabel('Available Seat Kilometers per Week')
plt.ylabel('Frequency')
plt.grid(False)
plt.show()
```

```python
# 2. Scatter Plot for Fatal Accidents (1985–1999) vs. (2000–2014)
plt.figure(figsize=(10, 6))
plt.scatter(df['fatal_accidents_85_99'], df['fatal_accidents_00_14'],
            alpha=0.5)
plt.title('Fatal Accidents (1985–1999) vs (2000–2014)')
plt.xlabel('Fatal Accidents 1985–1999')
plt.ylabel('Fatal Accidents 2000–2014')
plt.grid(True)
plt.show()
```

```python
# 3. Bar Plot for Total Fatalities per Airline
plt.figure(figsize=(10, 6))
fatalities = (df.groupby('airline')[['fatalities_85_99', 'fatalities_00_14']]
              .sum().sum(axis=1).sort_values(ascending=False))
plt.title('Total Fatalities per Airline')
plt.xlabel('Airline')
plt.ylabel('Total Fatalities')
plt.grid(False)
plt.show()
```
